```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

data = pd.read_excel("mp3_data.xlsx")

if "sra" in data.columns:
    data = data.drop(columns=["sra"])

#GRAFK 1: Platformlara Gre Ortalama Ekran Sresi
plt.figure()
sns.barplot(x="platform", y="Ekran sresi (saat)", hue="platform", legend=False, data=data, errorbar=None, palette="viridis")
plt.title("Platformlara Gre Ortalama Ekran Sresi")
plt.xlabel("Platform")
plt.ylabel("Ortalama Sre (Saat)")
plt.show()

#GRAFK 2: Platformlara Gre Ortalama Ya
plt.figure()
sns.barplot(x="platform", y="ya", hue="platform", legend=False, data=data, errorbar=None, palette="rocket")
plt.title("Platformlarn Kullanc Ya Ortalamas")
plt.xlabel("Platform")
plt.ylabel("Ortalama Ya")
plt.show()

#GRAFK 3: Zararl erik ve Ekran Sresi likisi
plt.figure()
sns.barplot(x="Maruz kalnan zararl ierik", y="Ekran sresi (saat)", hue="Maruz kalnan zararl ierik", legend=False, data=data, errorbar=None, palette="magma")
plt.title("Zararl erik Trne Gre Ortalama Ekran Sresi")
plt.xlabel("erik Tr")
plt.ylabel("Ortalama Sre (Saat)")
plt.xticks(rotation=90) 
plt.show()

#GRAFK 4: Korelasyon Matrisi (Heatmap)
plt.figure(figsize=(8, 6))
numeric_data = data.select_dtypes(include=['number']) 
corr = numeric_data.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Korelasyon Matrisi (Saysal likiler)")
plt.show()

if "sra" in data.columns:
    data = data.drop(columns=["sra"])

plt.figure()

sns.scatterplot(
    x="ya", 
    y="Ekran sresi (saat)", 
    hue="platform", 
    size="Ekran sresi (saat)", 
    sizes=(60, 300),
    alpha=0.8,
    data=data, 
    palette="deep"
)

plt.title("Hangi Ya Grubu Hangi Uygulamada Ne Kadar Vakit Harcyor?", fontsize=14, fontweight='bold')
plt.xlabel("Ya", fontsize=12)
plt.ylabel("Ekran Sresi (Saat)", fontsize=12)

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, title="Platform & Sre")

plt.tight_layout()
plt.show()
```
